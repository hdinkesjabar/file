```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel('/content/Puskesmas.xlsx')
```

```python
df.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
jumlah_kab = df['Kabkota'].value_counts()
jumlah_kab
```

```python
df_49 = df[df['1_Minggu_Sebelumnya'] < 0.50]
```

```python
df_49.tail()
```

```python
jml_kab_49 = df_49['Kabkota'].value_counts()
jml_kab_49
```

```python
jml_kab_49.plot(kind='pie', title='Persentase Kabupaten/kota Dengan Capaian < 50%', autopct='%0.2f%%', subplots=True, ylabel=' ', startangle=45)
plt.show()
```

```python
df_0 = df[df['1_Minggu_Sebelumnya'] == 0]
```

```python
jml_kab_0 = df_0['Kabkota'].value_counts()
jml_kab_0
```

```python
jml_kab_0.plot(kind='pie', title='Persentase Kabupaten/kota Dengan Capaian 0%', autopct='%0.2f%%', subplots=True, ylabel=' ', startangle=45)
plt.show()
```

```python
df_0['Kabkota'].value_counts().plot(kind='bar', title='Jumlah Kabupaten/kota Dengan Capaian 0%')
plt.show()
```

```python
df_49['Kabkota'].value_counts().plot(kind='bar', title='Jumlah Kabupaten/kota Dengan Capaian < 50%')
plt.show()
```
