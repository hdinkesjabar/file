```python
folder_path = "/home/buddhi/Downloads/Data Sehat/kabkot/"  # ganti sesuai folder Anda

# List untuk menampung dataframe tiap file
all_data = []

for file in os.listdir(folder_path):
    if file.endswith(".xlsx"):
        file_path = os.path.join(folder_path, file)

        # Baca data
        df = pd.read_excel(file_path)

        # Mapping nama kolom
        new_columns = {}
        for col in df.columns:
            if col.startswith("jumlah"):
                new_columns[col] = "Jumlah"
            elif col != "Jenis":
                new_columns[col] = col.replace("_", " ").title()

        df = df.rename(columns=new_columns)
        
        # Ubah semua value teks jadi kapital di awal kata
        for col in df.select_dtypes(include=["object"]).columns:
            df[col] = df[col].astype(str).str.title()

        # Tambahkan kolom 'Jenis' sesuai nama file (tanpa .xlsx)
        jenis_value = os.path.splitext(file)[0]
        df["Jenis"] = jenis_value

        # Masukkan ke list
        all_data.append(df)
```

```python
for i, df in enumerate(all_data):
    if df.columns.duplicated().any():
        print(f"⚠️ DataFrame ke-{i} punya kolom duplikat:", df.columns[df.columns.duplicated()].tolist())
```

```python
all_data = [
    df.loc[:, ~df.columns.duplicated()].copy().reset_index(drop=True)
    for df in all_data
]
```

```python
final_df_kabkot = pd.concat(all_data, ignore_index=True, sort=False)
final_df_kabkot
```

```python
# Simpan ke file Excel gabungan
output_file = os.path.join("/home/buddhi/Downloads/Data Sehat/", "Data Kab Kot.xlsx")
final_df_kabkot.to_excel(output_file, index=False)

print(f"Semua file sudah digabung → {output_file}")
```

```python
# Folder tempat file-file Excel disimpan
folder_path_jk = "/home/buddhi/Downloads/Data Sehat/jk/"  # ganti sesuai folder Anda

# List untuk menampung dataframe tiap file
all_data_jk = []

for file in os.listdir(folder_path_jk):
    if file.endswith(".xlsx"):
        file_path_jk = os.path.join(folder_path_jk, file)

        # Baca data
        df_jk = pd.read_excel(file_path_jk)

        # Mapping nama kolom
        new_columns_jk = {}
        for col in df_jk.columns:
            if col.startswith("jumlah"):
                new_columns_jk[col] = "Jumlah"
            elif col != "Jenis":
                new_columns_jk[col] = col.replace("_", " ").title()

        df_jk = df_jk.rename(columns=new_columns_jk)
        
        # Ubah semua value teks jadi kapital di awal kata
        for col in df_jk.select_dtypes(include=["object"]).columns:
            df_jk[col] = df_jk[col].astype(str).str.title()

        # Tambahkan kolom 'Jenis' sesuai nama file (tanpa .xlsx)
        jenis_value_jk = os.path.splitext(file)[0]
        df_jk["Jenis"] = jenis_value_jk

        # Masukkan ke list
        all_data_jk.append(df_jk)

# Gabungkan semua dataframe
final_df_jk = pd.concat(all_data_jk, ignore_index=True)

# Pastikan df sudah terbentuk
cols = list(final_df_jk.columns)

# Pindahkan 'Jenis Kelamin' ke paling kanan
if "Jenis Kelamin" in cols:
    cols.append(cols.pop(cols.index("Jenis Kelamin")))
    final_df_jk = final_df_jk[cols]

print(final_df_jk.columns)

final_df_jk
```

```python
output_file = os.path.join("/home/buddhi/Downloads/Data Sehat/", "Data JK.xlsx")
final_df_jk.to_excel(output_file, index=False)

print(f"Semua file sudah digabung → {output_file}")
```
