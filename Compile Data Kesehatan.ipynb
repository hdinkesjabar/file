## Menggabungkan Data Sekaligus Cleaning (Percobaan)

```python
# Folder tempat file-file Excel disimpan
folder_path = "/home/buddhi/Downloads/sampel"  # ganti sesuai folder Anda

# List untuk menampung dataframe tiap file
all_data = []

for file in os.listdir(folder_path):
    if file.endswith(".xlsx"):
        file_path = os.path.join(folder_path, file)

        # Baca data
        df = pd.read_excel(file_path)

        # Tambahkan kolom 'Jenis' sesuai nama file (tanpa .xlsx)
        jenis_value = os.path.splitext(file)[0]
        df["Jenis"] = jenis_value

        # Mapping nama kolom
        new_columns = {}
        for col in df.columns:
            if col.startswith("jumlah"):
                new_columns[col] = "Jumlah"
            elif col != "Jenis":
                new_columns[col] = col.replace("_", " ").title()

        df = df.rename(columns=new_columns)
        
        # Ubah semua value teks jadi kapital di awal kata
        for col in df.select_dtypes(include=["object"]).columns:
            df[col] = df[col].astype(str).str.title()

        # Masukkan ke list
        all_data.append(df)

# Gabungkan semua dataframe
final_df = pd.concat(all_data, ignore_index=True)

# Simpan ke file Excel gabungan
output_file = os.path.join(folder_path, "Data Kematian Gabungan.xlsx")
final_df.to_excel(output_file, index=False)

print(f"Semua file sudah digabung → {output_file}")
```
