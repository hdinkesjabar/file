#Menggabungkan Data dan Menambahkan Kolom Berdasarkan Nama File

## Import Library Yang Dibutuhkan

```python
import pandas as pd
import glob
import os
```

## Lihat Daftar File

```python
fff = glob.glob("/content/kate/*.xlsx")
fff
```

##Bersihkan File

```python
# Folder tempat file-file Excel disimpan
folder_path = "/content/kate/"  # ganti sesuai folder Anda

# List untuk menampung dataframe tiap file
all_data = []

for file in os.listdir(folder_path):
    if file.endswith(".xlsx"):
        file_path = os.path.join(folder_path, file)

        # Baca data
        df = pd.read_excel(file_path)

        # Mapping nama kolom
        # bila diawali jumlah_xxx dan kategori_xxx akan diubah jadi Jumlah dan Kategori
        new_columns = {}
        for col in df.columns:
            if col.startswith("jumlah"):
                new_columns[col] = "Jumlah"
            elif col.startswith("kategori"):
                new_columns[col] = "Kategori"
            elif col != "Jenis":
                new_columns[col] = col.replace("_", " ").title()

        df = df.rename(columns=new_columns)


        # Ubah semua value teks jadi kapital di awal kata
        for col in df.select_dtypes(include=["object"]).columns:
            df[col] = df[col].astype(str).str.title()

        # Tambahkan kolom 'Jenis' sesuai nama file (tanpa .xlsx)
        jenis_value = os.path.splitext(file)[0]
        df["Jenis"] = jenis_value

        # Masukkan ke list
        all_data.append(df)

# Gabungkan semua dataframe
final_df = pd.concat(all_data, ignore_index=True)
final_df.head()
```

##Bersihkan File Ke 2 pada folder JK

```python
# Folder tempat file-file Excel disimpan
folder_path2 = "/content/jk/"  # ganti sesuai folder Anda

# List untuk menampung dataframe tiap file
all_data2 = []

for fil in os.listdir(folder_path2):
    if fil.endswith(".xlsx"):
        file_path2 = os.path.join(folder_path2, fil)

        # Baca data
        df2 = pd.read_excel(file_path2)

        # Mapping nama kolom
        # bila diawali jumlah_xxx dan kategori_xxx akan diubah jadi Jumlah dan Kategori
        new_columns2 = {}
        for col2 in df2.columns:
            if col2.startswith("jumlah"):
                new_columns2[col2] = "Jumlah"
            elif col2.startswith("kategori"):
                new_columns2[col2] = "Kategori"
            elif col2 != "Jenis":
                new_columns2[col2] = col2.replace("_", " ").title()

        df2 = df2.rename(columns=new_columns2)


        # Ubah semua value teks jadi kapital di awal kata
        for col2 in df2.select_dtypes(include=["object"]).columns:
            df2[col2] = df2[col2].astype(str).str.title()

        # Tambahkan kolom 'Jenis' sesuai nama file (tanpa .xlsx)
        jenis_value2 = os.path.splitext(fil)[0]
        df2["Jenis"] = jenis_value2

        # Masukkan ke list
        all_data2.append(df2)

# Gabungkan semua dataframe
final_df2 = pd.concat(all_data2, ignore_index=True)

# Memastikan df sudah terbentuk
cols2 = list(final_df2.columns)

# Memindahkan kolom 'Jenis Kelamin' ke paling kanan
if "Jenis Kelamin" in cols2:
    cols2.append(cols2.pop(cols2.index("Jenis Kelamin")))
    final_df2 = final_df2[cols2]

final_df2.head()
```

## Menambahkan Kolom Jenis Kelamin dengan Value "-" pada Data Pertama (Yang Belum Ada Jenis Kelamin)

```python
# Menambahkan kolom jenis kelamin
final_df["Jenis Kelamin"] = "-"
final_df.head()
```

## Menggabungkan Data Pertama dan Kedua

```python
df_all = pd.concat([final_df2, final_df], ignore_index=True)
df_all.head()
```
